```python
#inizio pulizia dataset races
races = races.drop_duplicates() #rimozione di presunti duplicati
races = races.drop(columns=["time"]) #rimozione di colonne che non servono per l'analisi
races
```

```python
#filtro gli anni da tenere conto per la mia analisi
races = races[races["Year"] >= 2014]
races
```

```python
races.loc[:, "Date"] = pd.to_datetime(races["Date"], errors="coerce")
```

```python
races = races.rename(columns={"round": "Round"})
```

```python
#rimuovo i duplicati
```

```python
results = results.drop_duplicates() 
```

```python
#richiamo le colonne
```

```python
results.columns
```

```python
results = results.rename(columns={"race_id":"RaceID",
                                  "driver_id":"DriverID",
                                  "constructor_id":"ConstructorID",
                                  "grid":"Grid",
                                  "position":"Position",
                                  "position_order":"PositionOrder",
                                  "points":"Points",
                                  "laps":"Laps",
                                  "status":"Status"})
results
```

```python
#merge tra races e results per importare colonna Year e filtrare
results = results.merge(
    races[["RaceID", "Year"]],
    on="RaceID",
    how="left"
)
```

```python
#filtro per anno
results = results[results["Year"] >= 2014]
```

```python
#check
results
```

```python
#conversione tipo di dato alla colonna Year
results["Year"] = results['Year'].astype(int)
results
```

```python
results.info()
```

```python
#elimino la colonna Position perchè non ha tutti i valori numerici ed è uguale alla colonna PositionOrder che è totalmente numerica
results = results.drop(columns=["Position"], errors="ignore")
```

```python
#rinomino colonne per capire meglio
results = results.rename(columns={"Grid":"Starting_Position",
                                  "PositionOrder":"Final_Position"}
                        )
results
```

```python
#salvo le modifiche in un nouvo file csv pulito
results.to_csv("results_clean.csv", index=False)
```
